#Outliers are those values so far from the mean e.g most of the ages are from 10 - 30, we now have a few 100, 70, 80 and so on

```python
#To drop a column in the dataset.
dataset.drop('Cabin', axis = 1)
```

```python
#To drop a column in the dataset. Whatever drop you are doing, enforce that change on the original dataset.
dataset.drop('Cabin', axis = 1, inplace = True)
```

```python
dataset
```

```python
#If dropping more that one column, you put in a list:
#dataset.drop(['Cabin', 'PassengerId'], axis = 1, inplace = True)
```

```python
#How to fill null values
#Let's fill Age null values with the mean in the AGE COLUMN

dataset['Age'] = dataset['Age'].fillna(dataset['Age'].mean())
```

```python
#Now to check if Age still has null values

dataset.isnull().sum()
```

```python
#It is diffrent to fill a nonvalue category like the "Embarked' Column.
#Pick and use most occurent data/value which we have checked about 

dataset['Embarked'] = dataset['Embarked'].fillna('S')
```

```python
#now to check if the table still has null values

dataset.isnull().sum()
```

```python
dataset.info()
```

```python
#We need to change the 'AGE' data type to integer since it's float at the moment

dataset['Age'] = dataset['Age'].astype(int)
```

```python
dataset.info()
```

```python
#After importing our csv file, it changes to a DataFrame, now that we have cleaned it and we want to export the cleaned data, we have to change 
#back to a csv file. And you can save what you like

dataset.to_csv('Titanic Dataset_cleaned.csv')
```

## Data Visualization using Python 

The two visualization tools (libraries) used in python are:
 - Matplotlib
 - Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

 - Univariant - carrying out analysis with just one variable at a time e.g Histogram, Box plot (numeric value), Count plot (categories or objects)
 - Bivariate analysis - carrying out analysis between two variables e.g Line plot, Scatter plot, Bar Plot, Heatmap

## Box Plot - It shows the distribution of values and there are 5 summaries in that box plot

```python
plt.boxplot(dataset['Age'])
plt.show()
```

```python
#box plot for the distribution of 'Fare'

plt.boxplot(dataset['Fare'])
plt.show()
```

```python
#histogram 
# x is representing the age while y is respresenting the frequency in which the age appears
plt.hist(dataset['Age'])
plt.show()
```

```python
#To give the title of the axis and the chart title

plt.hist(dataset['Age'])
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
#For the 'Fare column

plt.hist(dataset['Fare'])
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Distribution of Fare')
plt.show()
```

```python
#To do frequency of values for categorical columns

sb.countplot(dataset['Sex'], color = 'r')
plt.xlabel('Frequency')
plt.title('Distribution of Sex')
```

```python
#using seaborn to plot the histogram chart

sb.histplot(dataset['Age'])
plt.show()
```

```python
dataset['Survived'].replace({0: 'No', 1: 'Yes'}, inplace = True)
```

```python
dataset
```

```python
sb.countplot(x='Sex', hue = 'Survived', data = dataset, palette = 'Set2')
plt.title('Total Survived By Sex')
```

```python
#To specify the colors you want
#You can aslo specify the color code

sb.countplot(x='Sex', hue = 'Survived', data = dataset, palette = ['b', 'r'])
plt.title('Total Survived By Sex')
```

```python
#To make the labels slant

sb.countplot(x='Sex', hue = 'Survived', data = dataset, palette = ['b', 'r'])
plt.title('Total Survived By Sex')
plt.xticks(rotation=45)
plt.show()
```

```python
#Survival by Passenger Class
#Countplot and catplot are different but they do the same thing

sb.catplot(x = 'Pclass', hue = 'Survived', data = dataset, kind = 'count', palette = 'pastel')
```

```python
#To resize the chart

plt.figure(figsize = (6,3))

sb.barplot(x = 'Pclass', y = 'Fare', data = dataset)
```

```python
#Scatter Plot - Between a numeric value and another numeric value

sb.scatterplot(x = 'Age', y = 'Fare', data = dataset)
plt.show
```
